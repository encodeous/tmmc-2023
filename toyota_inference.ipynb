```python
from ultralytics import YOLO
model = YOLO("yolo8m-ep150-dropout-0.01.pt")
classNames = ["bad", "good", "partial"]
```

```python
import math
import cv2

cap = cv2.VideoCapture(1)
cap.set(3, 640)
cap.set(4, 480)

# Assuming model is already loaded
# Assuming classNames is already defined

while True:
    success, img = cap.read()
    results = model(img, stream=True)

    # Initialize a dictionary to hold the counts of each class
    class_counts = {'good': 0, 'partial': 0, 'bad': 0}

    # Coordinates
    for r in results:
        boxes = r.boxes

        for box in boxes:
            # Bounding box
            x1, y1, x2, y2 = box.xyxy[0]
            x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2)  # convert to int values
            color_map = {"good" : (0, 255, 0), "partial" : (0, 255, 255), "bad" : (0, 0, 255)}

            # Put box in cam

            # Confidence
            confidence = math.ceil((box.conf[0]*100))/100

            # Class name
            cls = int(box.cls[0])
            class_name = classNames[cls]

            # Increment the count for the class
            class_counts[class_name] += 1

            cv2.rectangle(img, (x1, y1), (x2, y2), color_map[class_name], 3)
            # Object details
            org = [x1, y1]
            font = cv2.FONT_HERSHEY_SIMPLEX
            fontScale = 0.5
            color = (255, 0, 0)
            thickness = 1

            cv2.putText(img, class_name + str(confidence), org, font, fontScale, color, thickness)

    # Display the counts
    count_text = f"Good: {class_counts['good']} Partial Application: {class_counts['partial']} Bad: {class_counts['bad']}"
    cv2.putText(img, count_text, (10, 20), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 1)

    cv2.imshow('Webcam', img)
    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
